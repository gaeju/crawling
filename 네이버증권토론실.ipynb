```python
def get_day_list(item_code, page_no):
    url = f'https://finance.naver.com/item/board.naver?code={item_code}&page={page_no}'
    headers = {'user-agent':'Mozilla/5.0'}
    response = requests.get(url, headers = headers)

    table = pd.read_html(response.text)
    table = table[1]
    temp = table.drop('Unnamed: 6', axis = 1)
    temp = temp.dropna()
    return temp

def every_list(item_code, page_no):
    import time
    item_list = []


    while True:
        gdl = get_day_list(item_code, page_no)
        day = gdl.iloc[-1,0]
        item_list.append(gdl)

        if day <= '2023.01.10':
            break

        page_no += 1
        time.sleep(0.01)

    return pd.concat(item_list)

every_list('247540', 1)
```
