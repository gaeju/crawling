## 라이브러리 import

```python
import pandas as pd
import numpy as np
import requests
```

```python
# 수집할 url
url = "https://finance.naver.com/api/sise/etfItemList.nhn?etfType=0&targetColumn=market_sum&sortOrder=desc"
print(url)
```

## Pandas를 통한 요청

```python
result = pd.read_json(url, encoding="cp949").loc["etfItemList", "result"]
```

```python
pd.DataFrame(result)
```

## HTTP 요청

```python
response = requests.get(url)
```

```python
response.status_code
```

### JSON 타입으로 데이터 받기

```python
# requests 의 응답을 json 타입으로 받기
etf_json = response.json()
```

### JSON 에서 원하는 데이터 찾기

```python
# result > etfItemList 의 하위 구조로 목록을 찾고자 하는 데이터를 가져오기
etfItemList = etf_json['result']['etfItemList']
```

## 데이터프레임 변환

```python
df = pd.DataFrame(etfItemList)
df
```
