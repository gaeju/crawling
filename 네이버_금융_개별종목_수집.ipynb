## 페이지별 데이터 수집 함수 만들기

```python
# 종목 번호를 이용해 page에 따라 데이터를 읽어오는 함수
# 일자별 시세를 페이지별로 수집
def get_day_list(item_code, page_no):

  url = f'https://finance.naver.com/item/sise_day.naver?code={item_code}&page={page_no}'

  headers = {'user-agent': 'Mozilla/5.0'}
  response = requests.get(url, headers = headers)

  table = pd.read_html(response.text)

  temp = table[0].dropna()

  return temp

get_day_list("005930", 2)
```

```python
import time
# web page 시작번호
page_no = 1
item_code = '005930'
# 데이터를 저장할 빈 변수 선언
item_list = []
#while문을 이용하여 오늘~2021년 12월 1일까지
while True:
  #한 페이지를 수집하는 함수
  df_day = get_day_list(item_code, page_no)
  item_list.append(df_day)
  #마지막 행의 날짜 구하기
  day = df_day.iloc[-1,0]
  #날짜가 2021/1/1일 이전이라면 멈추기
  if day <= '2022.07.01':
    break

  time.sleep(0.01)
  page_no += 1

pd.concat(item_list)

```

## 전체 과정을 하나의 함수로 만들기

```python
def get_item_list(item_code, item_name):
#     """
#     일별 시세를 수집하는 함수
#     """
  page_no = 1
  # 데이터를 저장할 빈 변수 선언
  item_list = []
  #while문을 이용하여 오늘~2021년 12월 1일까지
  while True:
    #한 페이지를 수집하는 함수
    df_day = get_day_list(item_code, page_no)
    item_list.append(df_day)
    #마지막 행의 날짜 구하기
    day = df_day.iloc[-1,0]
    #날짜가 2021/1/1일 이전이라면 멈추기
    if day <= '2022.07.01':
      break

    time.sleep(0.01)
    page_no += 1

  df = pd.concat(item_list)

  df = df[df['날짜'] >= '2022.07.01']

  df['종목명'] = item_name
  df['종목코드'] = item_code

  cols = ['종목코드', '종목명', '날짜', '종가', '전일비', '시가', '고가', '저가', '거래량']
  df[cols]
  df.drop_duplicates


  date = df.iloc[0]['날짜']
  file_name = f'{item_code}-{item_name}-{day}.csv'
  df.to_csv(file_name, index = False)
  result = pd.read_csv(file_name)

  return result
```
